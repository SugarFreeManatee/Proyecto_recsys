## Creación set de test

Este código es para generar un dataset de test artificial en el que tengamos a disposición tanto playlist incompletas como las canciones que faltan (ground truth).

Para eso, descargamos el test set original del challenge, nos quedamos con las playlist que tengan más de una canción, y quitamos canciones de manera proporcional a la cantidad de canciones origindales:

- PL con 5 canciones -> quitamos 1 canción
- PL con 10 canciones -> quitamos 5 canciones
- PL con 25 canciones -> quitamos 10 canciones
- PL con 100 canciones -> quitamos 25 canciones

```python
# descargamos el challenge dataset (test) desde google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
folder_path = '/content/drive/MyDrive/Colab Notebooks/SR 2024/spotify_million_playlist_dataset_challenge'
```

```python
import pandas as pd

# el dataset viene en formato json
df = pd.read_json(folder_path + '/challenge_set.json')
print(df)
```

```python
df
```

```python
# nos quedamos solo con las columnas relevantes
df = df[["date", "playlists"]]
df
```

```python
# cada playlist dentro del df viene como json, las convertimos en columnas de un nuevo df
import json

json_columns = pd.json_normalize(df['playlists'])
```

```python
json_columns
```

```python
# debemos quedarnos solo con las playlist que tengan más de una canción
json_columns = json_columns[json_columns["num_samples"] > 1]
json_columns
```

```python
json_columns["num_samples"].unique()
```

```python
# agregamos una columna que tenga la playlist incompleta (la playlist de cada fila menos una cantidad x de canciones)
# y agregamos la columna con las canciones quitadas (las x canciones mencionadas anteriormente)
json_columns["tracks_incomplete"] = json_columns["tracks"].apply(lambda x: x[:len(x)-1] if len(x) == 5 else x[:len(x)-5] if len(x) == 10 else x[:len(x)-10] if len(x) == 25 else x[:len(x)-25])
json_columns["ground_truth"] = json_columns["tracks"].apply(lambda x: x[len(x)-1:] if len(x) == 5 else x[len(x)-5:] if len(x) == 10 else x[len(x)-10:] if len(x) == 25 else x[len(x)-25:])
```

```python
# agregamos una columna que diga cuántas canciones se quitaron de cada playlist
# y cuántas quedaron
json_columns["num_samples_final"] = json_columns["tracks_incomplete"].apply(lambda x: len(x))
json_columns["num_ground_truth_final"] = json_columns["ground_truth"].apply(lambda x: len(x))
```

```python
# dataset de test final
json_columns
```

```python
# convertimos el datset a csv para un mejor manejo futuro
json_columns.to_csv("test_set_final.csv")
```
